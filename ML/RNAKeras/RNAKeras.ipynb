```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.metrics import accuracy_score, f1_score, recall_score, confusion_matrix
from keras.models import Sequential # modelo sequencial que empilha as camadas
from keras.layers import Dense, Dropout # Dense = full connected, Dropout = exclui neuronios segundo uma probabilidade
```

```python
df = pd.read_csv("Churn_treino.csv", sep=";")
df
```

```python
x = df.drop("Exited", axis=1)
y = df["Exited"]
```

```python
# Padronizar os dados e fazer o LabelEncoder (categorias em numeros)
# Obs: precisa primeiro padronizar, se não padronizar o que foi
# gerado pelo LabelEncoder sempre dá ruim

standardscaler = StandardScaler()
numerical = x.select_dtypes(include=['int64', 'float64']).columns
x[numerical] = standardscaler.fit_transform(x[numerical])
```

```python
labelencoder = LabelEncoder()
categorical = x.select_dtypes(include='object').columns
for col in categorical:
    x[col] = labelencoder.fit_transform(x[col])
```

```python
x_treinamento, x_teste, y_treinamento, y_teste = train_test_split(x, y, test_size=0.3, random_state=0)
```

```python
modelo = Sequential()
modelo.add(Dense(units=64, activation='relu', input_dim=x_treinamento.shape[1])) # input_dim = numero de atributos do meu modelo
# quando pegamos o x e damos shape() a posição [0] é o número de linhas e a [1] é o número de colunas
modelo.add(Dropout(0.4))
modelo.add(Dense(units=32, activation='relu'))
modelo.add(Dropout(0.4))
modelo.add(Dense(units=64, activation='relu'))
modelo.add(Dropout(0.4))
modelo.add(Dense(units=1, activation='sigmoid'))
```

```python
modelo.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy']) # loss é como será feito o novo cálculo dos pesos
modelo.fit(x_treinamento, y_treinamento, epochs=50, batch_size=32)

```

```python
previsoes = modelo.predict(x_teste)
print(previsoes)
# Já que utilizamos sigmoid ele acaba não dando os valores em 0 e 1
# Mas ele dá os valores por faixas, então abaixo de 0.5 é 0 e a cima é 1
# Assim, faz mais sentido realizarmos uma transformação para facilitar a visualização
```

```python
y_pred = (previsoes > 0.5).astype('int32')
y_pred
```
