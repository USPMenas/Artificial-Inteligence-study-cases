```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.datasets import load_iris
from tpot import TPOTClassifier
```

```python
data = load_iris()
x = data.data
y = data.target
x_treinamento, x_teste, y_treinamento, y_teste = train_test_split(x, y, test_size=0.3, random_state=0)
```

```python
tpot = TPOTClassifier(
    generations = 10, # numero de geraçoes criadas
    population_size = 100, # quantos algoritmos, com quantas configuracoes vamos testar
    offspring_size = 100, # tamanho das gerações
    mutation_rate = 0.9, # probabilidade de algum elememento receber um valor aleatório
    crossover_rate = 0.1, # probabilidade de misturar as soluções
    scoring = 'accuracy', # métrica utilizada para escolher o modelo vencedor
    max_time_mins = 2, # tempo máximo do processo de otimização
    random_state = 0, 
    early_stop = False, # para o processo se não ocorrer evolução no mesmo
    verbosity = 2 # quantidade de informação que vamos gerar na saída
)

tpot.fit(x_treinamento, y_treinamento)
print("Melhor modelo: ", tpot.fitted_pipeline_)
```

```python
tpot.evaluated_individuals_
```

```python
previsao = tpot.predict(x_teste)
accuracy = accuracy_score(previsao, y_teste)
print("Accuracy: ", accuracy)
```
